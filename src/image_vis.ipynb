```python
images = []
labels = []
directory = f"../models/"
for model in os.listdir(directory):
    if model.startswith("2021"):
        path = os.path.join(directory, model)
        [date, time, data_name, blck, fct] = path.split("_")
        model_key = f"{date}_{time}"        
        for i in range(10):
            label = f"{data_name}_{blck}_{fct}"
            labels.append(label)
            image = f"{path}/image_0/image_0_{i}.png"
            images.append(image)
```

```python
ipyplot.plot_class_representations(images, labels, img_width=150)
```
